```
from transformers import pipeline
import gradio as gr

# --- 1. Define the Model and Pipeline (Replace with your actual model) ---
# This is a placeholder. You'll likely need a custom model trained for this task.
# For demonstration, we'll use a text generation model and adapt its output.

# In a real-world scenario, you might:
# - Fine-tune a language model (like BERT, RoBERTa, etc.) for course-specific recommendations.
# - Use a retrieval-based system with a database of courses and resources.

model_name = "gpt2"  # Replace with your actual model name or path
generator = pipeline('text-generation', model=model_name)

# --- 2. Function to Generate Book/Resource Suggestions ---
def suggest_resources(course_department):
    """
    Generates book and resource suggestions based on the student's course/department.

    Args:
        course_department (str): The name of the student's course or department.

    Returns:
        str: A string containing the suggested books and resources.
    """
    prompt = f"Based on the course/department '{course_department}', suggest some relevant books and learning resources:"
    generated_text = generator(prompt, max_length=200, num_return_sequences=1, pad_token_id=generator.tokenizer.eos_token_id)[0]['generated_text']

    # --- Post-process the generated text to extract suggestions ---
    # This is a very basic example. A more sophisticated approach might involve:
    # - Training the model to output in a specific format.
    # - Using regular expressions or NLP techniques to extract relevant information.
    suggestions = generated_text.split(prompt)[-1].strip()

    if not suggestions:
        return "Sorry, I couldn't find specific suggestions for that course/department at the moment."
    else:
        return f"Here are some suggested books and resources for '{course_department}':\n\n{suggestions}"

# --- 3. Create the Gradio Interface ---
iface = gr.Interface(
    fn=suggest_resources,
    inputs=gr.Textbox(label="Enter your Course or Department:"),
    outputs=gr.Textbox(label="Suggested Books and Resources:"),
    title="Course Resource Suggester",
    description="Enter your course or department to get suggestions for relevant books and learning materials."
)

# --- 4. Launch the Gradio Interface ---
iface.launch()
```
